### 1. Selection Sort ::

```python
def selection(arr):
  for i in range(0,len(arr)):
    si = i
    for j in range(i+1,len(arr)):
      if arr[j] < arr[si]:
        si = j
    arr[i], arr[si] = arr[si],arr[i]
arr = [2,9,3,4,5,6,0]
selection(arr)
print(arr)
```

### 2. Insertion Sort ::

```python
def insertion(arr):
  for i in range(1, len(arr)):
    ce = arr[i]
    j = i- 1
    while j >= 0 and ce <arr[j] :
      arr[j+1] = arr[j]
      j = j-1
    arr[j+1] = ce
```

```python
arr = [5,1,4,9,3,2,]
insertion(arr)
print(arr)
```

### 3. Bubble Sort ::

```python
def bubble(arr):
  for i in range(0,len(arr)-1):
    swap = False
    for j in range(len(arr)-i-1):
      if arr[j] > arr[j+1]:
        arr[j], arr[j+1] = arr[j+1], arr[j]
        swap = True
    if swap == False :
      break
```

```python
arr = [9,3,5,6,1]
bubble(arr)
print(arr)
```



### 4. Merge Sort :

```python
# merging part :
def merge_arr(left, right):
  result = []
  i , j = 0, 0
  n , m = len(left), len(right)

  while i <n and j < m :
    if left[i] <= right[j] :
      result.append(left[i])
      i+=1
    else :
      result.append(right[j])
      j+=1
  if i < n :
    while i < n :
      result.append(left[i])
      i+=1
  if j < n :
    while j < m:
      result.append(right[j])
      j+=1
  return result
```

```python
# divide the array into the left and right sub parts :
def merge_sort(arr):
  if len(arr)<=1 :
    return arr
  mid = len(arr)//2
  left_arr = arr[:mid]
  right_arr = arr[mid:]
  left = merge_sort(left_arr)
  right = merge_sort(right_arr)
  return merge_arr(left,right)
```

```python
arr = [9,4,5,6,77,6,5,4,4]
```

```python
r = merge_sort(arr)
print(r)
```

### 5. Quick Sort ::

```python
def partitional(arr,low,high):
  i ,j = low , high
  pivot = arr[low]
  while i<j :
    while arr[i]<=pivot and i<= high-1 :
      i +=1
    while arr[j] >= pivot and j >= low+1:
      j = j-1
    if i < j :
      arr[i],arr[j] = arr[j],arr[i]
    arr[low],arr[j] = arr[j],arr[low]
    return j
```

```python
def quicksort(arr, low, high):
  if low<high:
    pivot_index = partitional(arr,low,high)
    quicksort(arr,low,pivot_index-1)
    quicksort(arr,pivot_index+1,high)
```

```python
arr = [9,5,6,6,3,2,1]
n = len(arr)
quicksort(arr,0,n-1)
print(arr)
```
